```python
from pycoingecko import CoinGeckoAPI
from datetime import date
from datetime import time
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

cg = CoinGeckoAPI()
```

```python
now = datetime.now()
```

```python
now = datetime.now()
if now.minute >= 30:
    base = datetime.combine(date.today(),time(now.hour, 30))
    base = int(base.timestamp())-86400
else: 
    base = datetime.combine(date.today(),time(now.hour, 0))
    base = int(base.timestamp())-86400+1800

now = int(now.timestamp())

prices = cg.get_coin_market_chart_range_by_id(id='bitcoin',vs_currency='usd',from_timestamp=base,to_timestamp=now)
```

```python
x = []
y = []
for price in prices['prices']:
    x.append(price[0])
    y.append(price[1])
```

```python
def thousands(x, pos):
    'The two args are the value and tick position'
    return '$%1.1fK' % (x*1e-3)

formatter = FuncFormatter(thousands)

fig, ax = plt.subplots(figsize=(1.9, .3),dpi=130)
ax.yaxis.set_major_formatter(formatter)
ax.tick_params(axis="y", labelsize=6)
plt.plot(x, y)
plt.xticks([])
plt.show()
fig.savefig('test.png')
```

```python
for price in prices['prices']:
    if price[0]/1000 > (base + 1800):
        base = base + 1800
        candle_data[base] = {}
    
    if 'o' not in candle_data[base]:
        candle_data[base]['o'] = price[1]
    if 'h' not in candle_data[base]:
        candle_data[base]['h'] = price[1]
    if 'l' not in candle_data[base]:
        candle_data[base]['l'] = price[1]
    if 'c' not in candle_data[base]:
        candle_data[base]['c'] = price[1]
   
    if price[1] > candle_data[base]['h']:
        candle_data[base]['h'] = price[1]
    
    if price[1] < candle_data[base]['l']:
        candle_data[base]['l'] = price[1]
    
    candle_data[base]['c'] = price[1]
```
